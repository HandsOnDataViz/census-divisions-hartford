# Map of census boundaries for Hartford area
Includes county, county subdivision, census tract, census block

```python
import folium
```

```python
m = folium.Map(
    location=[41.763710, -72.685097],
    tiles=None,
    zoom_start=13,
    scrollWheelZoom=False
)

folium.TileLayer(
    tiles='cartodbpositron', 
    control=False
).add_to(m)

navy = 'rgb(0,0,51)'
blue = 'rgb(0,0,244)'
purple = 'rgb(143,5,250)'
pink = 'rgb(255,101,154)'
orange = 'rgb(255,174,81)'
yellow = 'rgb(255, 222, 22)'


def color_ct(a):
    return {
        'fillOpacity': 0,
        'weight': 11,
        #'color': 'purple'
        'color': yellow
    }

def color_counties(a):
    return {
        'fillOpacity': 0,
        'weight': 9,
        #'color': 'pink'
        'color': orange
    }

def color_subdivisions(a):
    return {
        'fillOpacity': 0,
        'weight': 7,
        #'color': 'orange'
        'color': pink
    }

def color_tracts(a):
    return {
        'fillOpacity': 0,
        'weight': 5,
        #'color': 'blue'
        'color': purple
    }

def color_block_groups(a):
    return {
        'fillOpacity': 0,
        'weight': 3,
        #'color': 'red'
        'color': blue
    }

def color_blocks(a):
    return {
        'fillOpacity': 0,
        'weight': 1,
        #'color': 'green'
        'color': navy
    }

folium.GeoJson(
    './geojson/hartford-blocks.geojson',
    name=f'Census blocks (Hartford) <span style="color: {navy}"></span>',
    style_function=color_blocks
).add_to(m)

folium.GeoJson(
    './geojson/hartford-block-groups.geojson',
    name=f'Census block groups (Hartford) <span style="color: {blue}"></span>',
    style_function=color_block_groups
).add_to(m)

folium.GeoJson(
    './geojson/hartford-tracts.geojson',
    name=f'Census tracts (Hartford) <span style="color: {purple}"></span>',
    style_function=color_tracts
).add_to(m)


folium.GeoJson(
    './geojson/hartford-county-subdivisions.geojson',
    name=f'County subdivisions (Hartford County) <span style="color: {pink}"></span>',
    style_function=color_subdivisions
).add_to(m)


folium.GeoJson(
    './geojson/ct-counties.geojson',
    name=f'Counties (Conn.) <span style="color: {orange}"></span>',
    style_function=color_counties
).add_to(m)

folium.GeoJson(
    './geojson/ct.geojson',
    name=f'State (Conn.) <span style="color: {yellow}"></span>',
    style_function=color_ct
).add_to(m)

folium.LayerControl(collapsed=False).add_to(m)

m.save('index.html')
m
```
